# Best location to open an Italian Restaurant in Toronto

## Description of the problem

### The objective of this project is :
To find the most suitable location for an entrepreneur to open an Italian restaurant in Toronto. By using data science methods such as clustering, this project aims to provide solutions to answer the business question: In Toronto, if an entrepreneur wants to open an Italian restaurant, where should they consider opening it?

## Description of the data and how it will be used to solve the problem

### To solve this problem, the data bellow is need it.
•	List of neighborhoods in Toronto. <br>
•	Latitude and Longitude of these neighborhoods. <br>
•	Venue data related to Italian restaurants. <br>

### Extracting the data
•	Scrapping of Toronto neighborhoods via Wikipedia. <br>
•	Getting latitude and longitude data of these neighborhoods via geocoder package. <br>
•	Using Foursquare API to get venue data related to these neighborhoods. <br>
